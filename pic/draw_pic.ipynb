```python
import numpy as np
import matplotlib.pyplot as plt
import os
filename = 'pytorch_convnext'

val_a = np.load(f'./{filename}/val_a.npy')
val_l = np.load(f'./{filename}/val_l.npy')
train_l = np.load(f'./{filename}/train_l.npy')
train_a = np.load(f'./{filename}/train_a.npy')

filename = 'pytorch_resnet'

val_a2 = np.load(f'./{filename}/val_a.npy')
val_l2 = np.load(f'./{filename}/val_l.npy')
train_l2 = np.load(f'./{filename}/train_l.npy')
train_a2 = np.load(f'./{filename}/train_a.npy')

x = range(1,201)

plt.subplot(2,2,1)
plt.plot(x,val_a,x,val_a2)
plt.legend(['convnext','resnet'])
plt.title('val_acc')

plt.subplot(2,2,2)
plt.plot(x,val_l,x,val_l2)
plt.legend(['convnext','resnet'])
plt.title('val_loss')

plt.subplot(2,2,3)
plt.plot(x,train_a,x,train_a2)
plt.legend(['convnext','resnet'])
plt.title('train_acc')

plt.subplot(2,2,4)
plt.plot(x,train_l,x,train_l2)
plt.legend(['convnext','resnet'])
plt.title('train_loss')

plt.subplots_adjust(wspace=0.4, hspace=0.4)
```
